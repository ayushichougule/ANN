```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x1=(1,1,-1,-1)
x2=(1,-1,1,-1)
t=(1,-1,-1,-1)
w=np.ones(2)
b=1
theta=0.2
lr=1
```

```python
#weight update,bias update,activation function
def weight_update(w,t,x1,x2):
    w[0]=w[0]+(lr*t*x1)
    w[1]=w[1]+(lr*t*x2)

def bias_update(b,t):
    return b+(lr*t)

def activation(yin):
    if yin>=theta:
        return 1
    else:
        return -1
```

```python
epoch =0
while (epoch<=2):
    for i in range(4):
        yin=x1[i]*w[0]+x2[i]*w[1]+b
        ycal=activation(yin)
        if ycal!=t[i]:
            weight_update(w,t[i],x1[i],x2[i])
            bias_update(b,t[i])
        else:
            pass
    epoch+=1
        
```

```python
slope=-(w[0]/w[1])
c=-(b/w[1])
line=lambda x:slope*x+c
y1=np.array([line(x) for x in x1])
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x=x1, y=x2, color="red")
plt.plot(np.array(x1), y1)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Perceptron Decision Boundary")
plt.grid(True)
plt.show()
```
