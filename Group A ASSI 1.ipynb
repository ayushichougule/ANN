```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def tanh(x):
    return np.tanh(x)
def relu(x):
    return np.maximum(0,x)
def linear(x):
    return x
```

```python
x=np.linspace(-5,5,100)
```

```python
sigmoid_y=sigmoid(x)
tanh_y=tanh(x)
relu_y=relu(x)
linear_y=linear(x)
```

```python
plt.figure(figsize=(12,8))
```

```python
plt.subplot(2,3,1)
plt.plot(x,sigmoid_y,label='Sigmoid')
plt.title('Sigmoid')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,3,2)
plt.plot(x,tanh_y,label='Tanh')
plt.title('Tanh')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,3,3)
plt.plot(x,relu_y,label='relu')
plt.title('relu')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,3,4)
plt.plot(x,linear_y,label='linear')
plt.title('linear')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
